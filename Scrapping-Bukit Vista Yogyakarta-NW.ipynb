## Yogyakarta

```python
# Menemukan elemen dengan class yang diinginkan
containers = soup_website.findAll('div', attrs={"class":"item-body flex-grow-1"})
print(containers)
```

```python
title_yogyakarta = []
for container in containers:
    title = container.find('h2', attrs={"class":'item-title'}).text
    print(title)
    title_yogyakarta.append((title))
    
data_title_yogyakarta = pd.DataFrame(title_yogyakarta, columns=['Yogyakarta Vacation Rental'])
data_title_yogyakarta
```

```python
yogyakarta_title = []  # List untuk menyimpan judul

for page in range(1, 4):  # Mengiterasi melalui 3 halaman
    url = f'https://www.bukitvista.com/city/yogyakarta/page{page}'  # Ganti URL dengan URL sebenarnya
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    #code diatas ialah untuk multiple page
    
    #define container terluar
    div_containers = soup.findAll('div', attrs={'class': 'd-flex align-items-center h-100'})

    for container in div_containers:
        review = container.find('h2', attrs={'class': 'item-title'})
        if review:
            yogyakarta_title.append(review.text.strip())
    
# Membuat dataframe dari list judul
df_yogyakarta_title = pd.DataFrame({'Yogyakarta Vacation Rental': yogyakarta_title})
print('total properties yang didapat : ', len(df_yogyakarta_title)) #validation data
# Menampilkan dataframe
df_yogyakarta_title.head()
```

```python
data_address = []  # List untuk menyimpan judul

for page in range(1, 4):  # Mengiterasi melalui 3 halaman
    url = f'https://www.bukitvista.com/city/yogyakarta/page{page}'  # Ganti URL dengan URL sebenarnya
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_containers = soup.findAll('div', attrs={'class': 'd-flex align-items-center h-100'})

    for container in div_containers:
        address = container.find('address', attrs={"class":'item-address'})
        if address:
            data_address.append(address.text.strip())
    
# Membuat dataframe dari list judul
df_yogyakarta_address = pd.DataFrame({'alamat': data_address})
print('total properties yang didapat : ', len(df_yogyakarta_address))
# Menampilkan dataframe
df_yogyakarta_address.head()
```

```python
price_residental = []  # List untuk menyimpan judul

for page in range(1, 4):  # Mengiterasi melalui 3 halaman
    url = f'https://www.bukitvista.com/city/yogyakarta/page{page}'  # Ganti URL dengan URL sebenarnya
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_containers = soup.findAll('div', attrs={'class': 'd-flex align-items-center h-100'})

    for container in div_containers:
        price = container.find('li', attrs={"class":'item-price item-price-text'})
        if price:
            price_residental.append(price.text.strip())
    
# Membuat dataframe dari list judul
df_yogyakarta_price = pd.DataFrame({'harga per malam ($ USD)': price_residental})
print('total properties yang didapat : ', len(df_yogyakarta_price))
# Menampilkan dataframe
df_yogyakarta_price.head()
```

```python
bedroom_yogyakarta = []  # List untuk menyimpan judul

for page in range(1, 4):  # Mengiterasi melalui 3 halaman
    url = f'https://www.bukitvista.com/city/yogyakarta/page{page}'  # Ganti URL dengan URL sebenarnya
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_containers = soup.findAll('div', attrs={'class': 'd-flex align-items-center h-100'})

    for container in div_containers:
        bedroom = container.find('li', attrs={"class": 'h-beds'})
        if bedroom:
            # Temukan elemen <span class="hz-figure"> dalam elemen 'bedroom'
            figure = bedroom.find('span', class_='hz-figure')
            if figure:
                # Ambil teks dari elemen 'figure' dan tambahkan ke daftar 'bedroom_yogyakarta'
                bedroom_yogyakarta.append(figure.get_text())
            else:
                # Jika tidak ada elemen 'figure', tambahkan None ke daftar 'bedroom_yogyakarta'
                bedroom_yogyakarta.append(None)
        else:
            # Jika tidak ada elemen 'bedroom', tambahkan None ke daftar 'bedroom_yogyakarta'
            bedroom_yogyakarta.append(None)
        #print(bedroom_yogyakarta)
    
# Membuat dataframe dari list judul
df_yogyakarta_kamar = pd.DataFrame({'kamar': bedroom_yogyakarta})
print('total properties yang didapat : ', len(df_yogyakarta_kamar))
# Menampilkan dataframe
df_yogyakarta_kamar.head()
```

```python
bathroom_yogyakarta = []  # List untuk menyimpan judul

for page in range(1, 4):  # Mengiterasi melalui 3 halaman
    url = f'https://www.bukitvista.com/city/yogyakarta/page{page}'  # Ganti URL dengan URL sebenarnya
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_containers = soup.findAll('div', attrs={'class': 'd-flex align-items-center h-100'})

    for container in div_containers:
        bathroom = container.find('li', attrs={"class": 'h-baths'})
        if bathroom:
            # Temukan elemen <span class="hz-figure"> dalam elemen 'bathroom'
            figure = bathroom.find('span', class_='hz-figure')
            if figure:
                # Ambil teks dari elemen 'figure' dan tambahkan ke daftar 'bathroom_yogyakarta'
                bathroom_yogyakarta.append(figure.get_text())
            else:
                # Jika tidak ada elemen 'figure', tambahkan None ke daftar 'bathroom_yogyakarta'
                bathroom_yogyakarta.append(None)
        else:
            # Jika tidak ada elemen 'bathroom', tambahkan None ke daftar 'bathroom_yogyakarta'
            bathroom_yogyakarta.append(None)
    
# Membuat dataframe dari list judul
df_yogyakarta_kamarmandi = pd.DataFrame({'kamar mandi': bathroom_yogyakarta})
print('total properties yang didapat : ', len(df_yogyakarta_kamarmandi))
# Menampilkan dataframe
df_yogyakarta_kamarmandi.head()
```
